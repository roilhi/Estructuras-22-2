## Complejidad computacional
La eficiencia de un algoritmo computacional gira en torno a dos preguntas básicas:
* ¿Cuánto tiempo se requiere para que finalice este algoritmo?
* ¿Cuánto espacio en memoria se requiere para los cálculos de este algoritmo?

No es justo resolver el problema del tiempo mediante algún cronómetro, ya que el rendimiento debe ser independiente del hardware

¿Qué hacer para conocer la eficiencia o complejidad computacional de un algoritmo?
*  Contar el número de operaciones en tiempo fijo
*  Absoluta dependencia del tamaño de la entrada ($n\rightarrow\infty$, mayor tiempo)
*  Comúnmente el tamaño de la entrada se define como $n$ ó $N$
*  El algoritmo es eficiente si se ejecuta en un tiempo polinomial con respecto a la entrada (P)
*  En ocasiones, no existe un algoritmo eficiente para la entrada, debe estimarse o reducirse (NP, NP Complete, NP Hard)

## Análisis asintótico
*  Se ignoran los factores constantes y los términos de bajo orden
*  Se tienen 3 casos distintos: 

1.  Big $O$: Límite superior (peor caso)
2.  Big $\Omega$: Límite inferior (la más eficiente de las soluciones)
3.  Big $\Theta$: Intervalo entre ambos límites

A continuación se presenta un gráfico de algunos de los casos comunes de la notación Big-O

## Notación Big-O
* Como se basa en el límite superior de la complejidad, brinda el peor escenario
* Cuantifica el rendimiento mientras el tamaño de la entrada se vuelve arbitrariamente grande

```python
import numpy as np
import matplotlib.pyplot as plt

# Estilo de hoja para matplotlib
plt.style.use('bmh')

# Crear un vector con los elementos n de 1 a 1000
n = np.linspace(1, 10, 1000)
# Etiquetas para la leyenda
labels = ['$O(1)$', '$O(\log(n))$', '$O(n)$', '$O(n\log(n))$', '$O(n^2)$', '$O(n^3)$', '$O(2^n)$']
# Lista con los arreglos big O a graficar
big_o = [np.ones(n.shape), np.log(n), n, n * np.log(n), n**2, n**3, 2**n]
# Tamaño de gráfico y corte en eje Y (hasta 50)
plt.figure(figsize=(10, 8))
plt.ylim(0, 50)
# Iterando y graficando los elementos de la lista
for i in range(len(big_o)):
    plt.plot(n, big_o[i], label=labels[i])
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 15}

plt.rc('font', **font)
plt.legend(loc=0)
plt.title('Gráficos de complejidad en notación big-O')
plt.ylabel('Tiempo Relativo')
plt.xlabel('Tamaño de entrada $n$')
plt.savefig('NotacionBigO.png')
```

### Propiedades de la notación big-O
*  $O(f(n)) = O(n)$ si y solo si $f(n)$ es una función lineal
*  $O(cf(n)) = O(n)$
*  Si $h=O(g)$ y $g=O(f)$ entonces $h=O(f)$
*  Si $g_1 \in O(f_1)$ y $g_2 \in O(f_2)$ entonces $g_1+g_2 \in O(\max \{f_1,f_2\})$
*  Si $g_1 \in O(f_1)$ y $g_2 \in O(f_2)$ entonces $g_1g_2 \in O(f_1 f_2)$
*  Si $g_1 \in O(f)$ y $g_2 \in O(f^k)$ entonces $g_1g_2 \in O(f^{k+1})$

Sea $f$ una función que describe el tiempo de ejecución (complejidad) de un algoritmo en particular para una entrada tamaño $n$:

*  $f(n) = 7\log(n)^3 + 15n^2 + 12n^3 +8 $
*  $O(f(n)) = O(n^3) $

### Complejidad constante $O(1)$
Cualquier operación aritmética o asignación de variables/constantes independiente del tamaño de la entrada $n$ tiene una complejidad constante $O(1)$

```python
# Ejemplo O(1)
a = 1
b = 2
c = a + 5 * b
```

```python
# Ejemplo 2 O(1)
i = 0
while i < 11:
    i + = 1
```

### Complejidad de tiempo lineal $O(n)$
Operaciones de búsqueda lineal o recorrido lineal de un array de tamaño $n$ o iteraciones de bucles sobre ese array

```python
# Ejemplo O(f(n)) = O(n)
# f(n) = n
i = 0
while i < n:
    i + = 1
    f(n) = n
```

```python
# Ejemplo O(f(n)) = O(n)
# f(n) = n/3
i = 0
while i < n:
    i + = 3
```

### Complejidad de tiempo cuadrático $O(n^2)$
Se presenta de manera muy común al iterar sobre un array en dos ciclos anidados


```python
# f(n) = n*n= n^2 -- O(f(n)) = O(n^2)
for i in range(1,n):
    for j in range(1,n):
        print(n)
```

En el siguiente caso, el intervalo del ciclo interno de la variable $j$ llegue a acotarse por el tamaño de $i$. Dado que $i \in [0,n)$ se tienen los siguientes casos:
*  Si $i=0$, entonces $j=1,2,\cdots n$
*  Si $i=1$, entonces $j=1,2,\cdots (n-1)$
*  Si $i=2$, entonces $j=1,2,\cdots (n-2)$

De esta manera, al agrupar todos los casos se tiene que
$ (n) + (n-1) + (n-2) + \cdots +2+1 = \frac{n(n+1)}{2} \rightarrow O(n^2)$

```python
# f(n) = n*n= n^2 -- O(f(n)) = O(n^2)
for i in range(1,n):
    for j in range(i,n):
        print(n)
```
